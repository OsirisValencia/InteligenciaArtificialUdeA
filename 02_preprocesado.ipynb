```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualizar_columnas_con_pocos_valores(dataframe, max_valores=13):
    # Identificar columnas con menos de max_valores valores únicos
    columnas_filtradas = [col for col in dataframe.columns if dataframe[col].nunique() < max_valores]

    if not columnas_filtradas:
        print(f"No hay columnas con menos de {max_valores} valores únicos.")
        return

    print(f"Se encontraron {len(columnas_filtradas)} columnas con menos de {max_valores} valores únicos.\n")

    # Para cada columna, crear una visualización
    for i, columna in enumerate(columnas_filtradas):
        # Obtener conteo de valores
        conteo = dataframe[columna].value_counts(dropna=False)

        df_conteo = pd.DataFrame({
            'Valor': conteo.index.astype(str),
            'Cantidad': conteo.values,
            'Porcentaje': (conteo.values / len(dataframe) * 100).round(2)

        })

        # Ordenar por cantidad (mayor a menor)
        df_conteo = df_conteo.sort_values('Cantidad', ascending=False).reset_index(drop=True)

        # Imprimir tabla de distribución
        print(f"\n--- Columna: {columna} ({conteo.count()} valores únicos) ---")
        print(df_conteo.to_string(index=False))


# Ejemplo de uso
visualizar_columnas_con_pocos_valores(df)
```

# Proceso de limpieza correspondiente.

Primero debemos definir qué tipo de dato tienen las columnas, si es un tipo de dato categorico, discreto, continuo...
Esto nos ayudará a definir una estrategía para saber cómo rellenar los valores faltantes.
Inicialmente el dataset todos los tipos de datos son objects, vamos a "Normalizarlos" siguiendo los diferentes tipos de datos. por ejemplo algunos solo tienen 2 posibles valores si y no, los trataremos como binarios. hay algunos valores discretos como lo puede ser ESTRATO_VIVIENDA y otros continuos como lo puede ser valormatriculauniversidad que a su vez puede ser categorico.
<br/>
Anteriormente identificamos las siguientes columnas con valores faltantes:




| Tipo                     | Columnas                                                                                                                                                         | Nº categorías/distintos |
| ------------------------ | ---------------------------------------------------------------------------------------------------------------------------------------------------------------- | ----------------------- |
| **Continuas**            | coef\_1, coef\_2, coef\_3, coef\_4                                                                                                                               | 375, 342, 196, 276      |
| **Ordinales**            | ESTU\_VALORMATRICULAUNIVERSIDAD (8+1 nan), ESTU\_HORASSEMANATRABAJA (5+1 nan), FAMI\_ESTRATOVIVIENDA (7+1 nan), PERIODO (9 valores, pero solo **primeros 4 dígitos** relevantes: año + posible semestre)                                                   | 8, 5, 7, 4                 |
| **Binarias (`Si`/`No`)** | FAMI\_TIENEINTERNET, FAMI\_TIENELAVADORA, FAMI\_TIENEAUTOMOVIL, ESTU\_PRIVADO\_LIBERTAD, ESTU\_PAGOMATRICULAPROPIO, FAMI\_TIENECOMPUTADOR, FAMI\_TIENEINTERNET.1 | todas tienen 2+1 nan    |
| **Multiclase altas**     | ESTU\_PRGM\_ACADEMICO (948), ESTU\_PRGM\_DEPARTAMENTO (31), FAMI\_EDUCACIONPADRE (12+1 nan), FAMI\_EDUCACIONMADRE (12+1 nan)                                     | 948, 31, 12, 12         |                                                                       | 9                       |
| **Target**               | RENDIMIENTO\_GLOBAL (4 clases balanceadas)                                                                                                                       | 4                       |


**Tratamiento de PERIODO** Cada valor es un entero de 5 dígitos como 20195, pero solo nos interesa los 4 primeros indican año Propuesta:

Convertir a string y quedarnos con los 4 primeros caracteres:

*Estrategias propuestas para la inputacion de datos según el tipo de datos que haya en la columna*

| Tipo de variable           | Estrategia recomendada                                                                                                                                  |
| -------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Continuas**              | – Imputación por **mediana** (robusta frente a outliers).<br>– Alternativa: KNNImputer si hay correlaciones.                                            |
| **Ordinales**              | – Imputar con **valor modal** (la categoría más frecuente).<br>– O crear categoría “Desconocido” para conservar missing.                                |
| **Binarias (`Si`/`No`)**   | – Imputar con la **frecuencia** de `Si`/`No` (modo).<br>– O codificar `nan` como categoría propia (`-1` o `Unknown`).                                   |
| **Multiclase altas**       | – Dado el alto cardinal, imputar con categoría “Desconocido” para no distorsionar.<br>– O usar técnicas basadas en similitud (e.g. MostFrequent / KNN). |         |


| Tipo                                              | Columnas & estrategia                                                              |
| ------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **Continuas**                                     | – **StandardScaler** (media=0, var=1) para coef\_\*.                               |
| **Ordinales**                                     | – **OrdinalEncoder** respetando el orden natural:<br>  - ESTU\_HORASSEMANATRABAJA: |
| `["0","Menos de 10","11-20","21-30","Más de 30"]` |                                                                                    |


```python
df['PERIODO_STR'] = df['PERIODO'].astype(str).str.zfill(5)
df['SEMESTRE'] = df['PERIODO_STR'].str[:4].astype(int)
df = df.drop(['PERIODO','PERIODO_STR'], axis=1)
```

```python
from sklearn.compose        import ColumnTransformer
from sklearn.impute         import SimpleImputer, KNNImputer
from sklearn.preprocessing  import StandardScaler, OneHotEncoder, OrdinalEncoder


# Definición de columnas
numeric_cols    = ["coef_1","coef_2","coef_3","coef_4"]
ordinal_cols    = ["SEMESTRE",
                   "ESTU_HORASSEMANATRABAJA",
                   "FAMI_ESTRATOVIVIENDA",
                   "ESTU_VALORMATRICULAUNIVERSIDAD"]
binary_cols     = ["FAMI_TIENEINTERNET", "FAMI_TIENELAVADORA",
                   "FAMI_TIENEAUTOMOVIL","ESTU_PRIVADO_LIBERTAD",
                   "ESTU_PAGOMATRICULAPROPIO","FAMI_TIENECOMPUTADOR"]
highcard_cols   = ["FAMI_EDUCACIONPADRE","FAMI_EDUCACIONMADRE"]  # excluyendo el programa y la región

# Imputación de valores faltantes
#    - Numéricas: mediana
num_imp = SimpleImputer(strategy='median')
df[numeric_cols] = num_imp.fit_transform(df[numeric_cols])

#  3)- Ordinales: moda (most_frequent), categorizando nan como 'Desconocido'
ord_imp = SimpleImputer(strategy='most_frequent', fill_value='Desconocido')
df[ordinal_cols] = ord_imp.fit_transform(df[ordinal_cols])

#  4) Binarias: moda (most_frequent), categorizando nan como 'Desconocido'
bin_imp = SimpleImputer(strategy='most_frequent', fill_value='Desconocido')
df[binary_cols] = bin_imp.fit_transform(df[binary_cols])

# 5) Columnas de alta cardinalidad
high_imp = SimpleImputer(strategy='constant', fill_value='Desconocido')

# Ajuste y transformación
df[highcard_cols] = high_imp.fit_transform(df[highcard_cols])

# 5) Mostrar resultados
print("Primeras filas tras imputación:")
print(df.head())

print("\nConteo de valores nulos por columna:")
print(df.isnull().sum())

print("\nDistribución de categorías (incluyendo 'Desconocido'):")
for col in highcard_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False).head(10))
```

```python
from pandas.api.types import CategoricalDtype

# Definir los órdenes para variables ordinales
horas_order = ["0", "Menos de 10 horas", "Entre 11 y 20 horas",
               "Entre 21 y 30 horas", "Más de 30 horas", "Desconocido"]
estrato_order = ["Estrato 1", "Estrato 2", "Estrato 3",
                 "Estrato 4", "Estrato 5", "Estrato 6", "Sin Estrato", "Desconocido"]
matri_order = ["Menos de 500 mil", "Entre 500 mil y menos de 1 millón",
               "Entre 1 millón y menos de 2.5 millones",
               "Entre 2.5 millones y menos de 4 millones",
               "Entre 4 millones y menos de 5.5 millones",
               "Entre 5.5 millones y menos de 7 millones",
               "Más de 7 millones", "No pagó matrícula", "Desconocido"]

# 2) Crear CategoricalDtype para cada ordinal
ord_horas = CategoricalDtype(categories=horas_order, ordered=True)
ord_estrato = CategoricalDtype(categories=estrato_order, ordered=True)
ord_matri = CategoricalDtype(categories=matri_order, ordered=True)

# 3) Asignar dtypes
df = df.astype({
    'FAMI_EDUCACIONPADRE': 'category',
    'FAMI_EDUCACIONMADRE': 'category',
    'FAMI_TIENEINTERNET': 'category',
    'FAMI_TIENELAVADORA': 'category',
    'FAMI_TIENEAUTOMOVIL': 'category',
    'ESTU_PRIVADO_LIBERTAD': 'category',
    'ESTU_PAGOMATRICULAPROPIO': 'category',
    'FAMI_TIENECOMPUTADOR': 'category',
    'SEMESTRE': 'int16',
    'coef_1': 'float64',
    'coef_2': 'float64',
    'coef_3': 'float64',
    'coef_4': 'float64',
})

# 4) Aplicar categorías ordenadas
df['ESTU_HORASSEMANATRABAJA'] = df['ESTU_HORASSEMANATRABAJA'].astype(ord_horas)
df['FAMI_ESTRATOVIVIENDA'] = df['FAMI_ESTRATOVIVIENDA'].astype(ord_estrato)
df['ESTU_VALORMATRICULAUNIVERSIDAD'] = df['ESTU_VALORMATRICULAUNIVERSIDAD'].astype(ord_matri)

# 5) Verificación
print(df.dtypes)
```

```python
import unicodedata, re
def normalize_text(s):
    s = str(s).lower()
    s = unicodedata.normalize('NFKD', s).encode('ASCII', 'ignore').decode('ASCII')
    s = re.sub(r'[^a-z0-9 ]+', ' ', s)
    return re.sub(r'\s+', ' ', s).strip()

df['PRGM_NORM'] = df['ESTU_PRGM_ACADEMICO'].map(normalize_text)
```

```python
!pip install rapidfuzz
```

```python
from rapidfuzz import fuzz
from scipy.cluster.hierarchy import linkage, fcluster

# 1) lista única
progs = df['PRGM_NORM'].unique().tolist()

# 2) matriz de similitud (sparse si es necesario)
N = len(progs)
D = np.zeros((N, N))
for i in range(N):
    for j in range(i+1, N):
        score = fuzz.token_sort_ratio(progs[i], progs[j]) / 100
        D[i,j] = D[j,i] = 1 - score

# 3) linkage y corte
Z = linkage(D, method='average')
labels = fcluster(Z, t=0.3, criterion='distance')  # t ajusta sensibilidad

# 4) mapping de programa → clúster
cluster_map = dict(zip(progs, labels))
df['PRGM_CLUSTER'] = df['PRGM_NORM'].map(cluster_map)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans

progs = df['PRGM_NORM'].unique().tolist()

model = SentenceTransformer('all-MiniLM-L6-v2')
embs  = model.encode(progs, show_progress_bar=True)
kmeans = KMeans(n_clusters=80).fit(embs)
cluster_map_sem = dict(zip(progs, kmeans.labels_))
df['PRGM_CLUSTER_SEM'] = df['PRGM_NORM'].map(cluster_map_sem)
```

Ahora vamos a normalizar las columnas faltantes para que luego sea posible en la siguiente entrega evaluarlas con algún modelo de clasificación.

```python
from sklearn.preprocessing import OrdinalEncoder

matri_order = [
  'Menos de 500 mil',            # 0
  'Entre 500 mil y menos de 1 millón',
  'Entre 1 millón y menos de 2.5 millones',
  'Entre 2.5 millones y menos de 4 millones',
  'Entre 4 millones y menos de 5.5 millones',
  'Entre 5.5 millones y menos de 7 millones',
  'Más de 7 millones',           # 6
  'No pagó matrícula',           # 7
  'Desconocido'                  # 8 (si aplica)
]

horas_order = [
  '0', 'Menos de 10 horas',
  'Entre 11 y 20 horas',
  'Entre 21 y 30 horas',
  'Más de 30 horas',
  'Desconocido'
]

estrato_order = [
  'Estrato 1','Estrato 2','Estrato 3',
  'Estrato 4','Estrato 5','Estrato 6',
  'Sin Estrato','Desconocido'
]

ord_enc_matri = OrdinalEncoder(
    categories=[matri_order], dtype=int,
    handle_unknown='use_encoded_value', unknown_value=-1
)
df['VALOR_MATRICULA_ENC'] = ord_enc_matri.fit_transform(
    df[['ESTU_VALORMATRICULAUNIVERSIDAD']]
)

ord_enc_horas = OrdinalEncoder(
    categories=[horas_order], dtype=int,
    handle_unknown='use_encoded_value', unknown_value=-1
)
df['HORAS_TRABAJO_ENC'] = ord_enc_horas.fit_transform(
    df[['ESTU_HORASSEMANATRABAJA']]
)

ord_enc_estrato = OrdinalEncoder(
    categories=[estrato_order], dtype=int,
    handle_unknown='use_encoded_value', unknown_value=-1
)
df['ESTRATO_ENC'] = ord_enc_estrato.fit_transform(
    df[['FAMI_ESTRATOVIVIENDA']]
)

# El Ordinal Encoding preserva el orden y consume solo una columna
# El Ordinal Encoding es óptimo para ordinales porque reduce dimensionalidad y mantiene el orden
```

```python
# Variables binarias


bin_map = {'No': 0, 'N': 0, 'Si': 1, 'S': 1}
bin_cols = [
  'FAMI_TIENEINTERNET','FAMI_TIENELAVADORA',
  'FAMI_TIENEAUTOMOVIL','ESTU_PRIVADO_LIBERTAD',
  'ESTU_PAGOMATRICULAPROPIO','FAMI_TIENECOMPUTADOR'
]
for col in bin_cols:
    df[col + '_ENC'] = df[col].map(bin_map).astype(int)

```

```python
edu_order = [
  'Ninguno',
  'Primaria incompleta','Primaria completa',
  'Secundaria (Bachillerato) incompleta','Secundaria (Bachillerato) completa',
  'Técnica o tecnológica incompleta','Técnica o tecnológica completa',
  'Educación profesional incompleta','Educación profesional completa',
  'Postgrado',
  'No Aplica','No sabe','Desconocido'
]
ord_enc_padre = OrdinalEncoder(
    categories=[edu_order], dtype=int,
    handle_unknown='use_encoded_value', unknown_value=-1
)
df['EDU_PADRE_ENC'] = ord_enc_padre.fit_transform(
    df[['FAMI_EDUCACIONPADRE']]
)
df['EDU_MADRE_ENC'] = ord_enc_padre.fit_transform(
    df[['FAMI_EDUCACIONMADRE']]
)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# 1) Definir y aplicar encoders ordinales
ord_encoders = {
  'VALOR_MATRICULA_ENC': ( ['ESTU_VALORMATRICULAUNIVERSIDAD'], matri_order ),
  'HORAS_TRABAJO_ENC':   ( ['ESTU_HORASSEMANATRABAJA'], horas_order ),
  'ESTRATO_ENC':         ( ['FAMI_ESTRATOVIVIENDA'], estrato_order ),
  'EDU_PADRE_ENC':       ( ['FAMI_EDUCACIONPADRE'], edu_order ),
  'EDU_MADRE_ENC':       ( ['FAMI_EDUCACIONMADRE'], edu_order )
}

for new_col, (cols, order) in ord_encoders.items():
    enc = OrdinalEncoder(
        categories=[order], dtype=int,
        handle_unknown='use_encoded_value', unknown_value=-1
    )
    df[new_col] = enc.fit_transform(df[cols])

# 2) Mapear binarios
for col in bin_cols:
    df[col + '_ENC'] = df[col].map(bin_map).astype(int)
```
